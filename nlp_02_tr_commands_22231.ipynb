# implementing tr commands

1. c is complement , so set1 is 'A-Za-z' and set2 is '\n'

```python
!tr -cs 'A-Za-z' '\n' < /content/drive/MyDrive/sss_01_01.txt
```

2. Sort the lines and pass them to uniq -c which will collapse

This will give the frequency of the token

```python
!tr -sc 'A-Za-z' '\n' < /content/drive/MyDrive/sss_01_02.txt | sort | uniq -c
```

3.  sorting numerically and in reverse order, so the token with more occurence will be first

```python
! tr -sc 'A-Za-z' '\n' < /content/drive/MyDrive/sss_01_03.txt | tr A-Z a-z | sort | uniq -c | sort -n -r
```
