```python
# Dictionary to store frequent adjacent pairs
pair_frequency = {} #dictionary


for token in filtered_tokens: # going through each word in the filtered_tokens list
    for i in range(len(token) - 1):   # going through all the characters of the token
        pair = (token[i], token[i + 1]) # taking each pair from the string
        if pair in pair_frequency: # if the pair is already there in the dictionary then increase its count by 1
            pair_frequency[pair] += 1
        else: # else the pair is not yet there in the dictionary , give its count value as 1
            pair_frequency[pair] = 1

print(pair_frequency)
```

```python
def byte_pair_encoding(corpus, num_merges): # function byte pair encoding which takes parameters corpus and number of merges
    vocab = set(' '.join(corpus))  # Initial set of tokens is characters

    for _ in range(num_merges): # iterating over the loop for number of merges times
        token_counts = {}

        # Count frequencies of adjacent token pairs
        for sentence in corpus:  # for each token in the corpus
            for j in range(len(sentence) - 1): # Going through all the characters of the string
                pair = sentence[j] + sentence[j + 1] # taking every adjacent pair(of characters) from the token
                if pair in token_counts: # if the pair is already in token_counts
                    token_counts[pair] += 1 # increase its count by 1
                else:
                    token_counts[pair] = 1 # means the pair is not present yet in the token_counts dictionary . we will add it into it now

        # Find the most frequent token pair
        most_common_pair = max(token_counts, key=token_counts.get) # we are getting the key of most common pair from the dictionary with max frequency

        # Create a new token by using the most common pair
        tNEW = most_common_pair # assigning the most_common_pair we got to the tNEW variable

        # Update the vocabulary
        vocab.add(tNEW) # adding the new variable to the vocabulary

        # Update the corpus by replacing occurrences of the most common pair
        new_corpus = [] # list called new_corpus
        for sentence in corpus: # for every word in the corpus
            new_sentence = sentence.replace(most_common_pair, tNEW) #if the pair is found in the corpus replace it as a new character in the entire corpus
            new_corpus.append(new_sentence) # adding the updated corpus into the enew_corpus
        corpus = new_corpus # assigning back to the variable corpus

    return vocab # returning the set vocalbulary

# Set the number of merges
num_merges = 10

# Apply Byte Pair Encoding
vocab = byte_pair_encoding(filtered_tokens, num_merges) # passing filtered_tokens(basically corpus), and number of merges
print(vocab) # printing vocabulary
```
